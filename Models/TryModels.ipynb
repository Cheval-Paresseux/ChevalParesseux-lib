```python
import sys
sys.path.append("../")
from Models import LinearRegression as reg

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#  standard dcaler 
from sklearn.preprocessing import StandardScaler
```

```python
X = pd.Series([i for i in range(100)])
y = 10 + 2*X + np.random.normal(0, 10, 100)

plt.figure(figsize=(17, 5))
plt.scatter(X, y)
plt.show()
```

```python
model = reg.MSERegression()
model.fit(X, y, learning_rate=0.1, epochs=1000)
coefficients = model.coefficients
intercept = model.intercept
statistics, _ = model.get_statistics()
print(f"Model coefficients: {coefficients}, Intercept : {intercept}")
print("R squared: ", statistics['R_squared'])

model.plot_loss_history()
```

```python
model = LinearRegression()
model.fit(X.values.reshape(-1, 1), y)
print(f"Model coefficients: {model.coef_}, Intercept : {model.intercept_}")
print("R squared: ", model.score(X.values.reshape(-1, 1), y))
```
