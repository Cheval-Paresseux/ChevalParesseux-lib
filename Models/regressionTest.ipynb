```python
import Regression as reg
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
beta = 200
alpha = 1000

x = np.random.randn(100)
y = alpha + beta*x + np.random.randn(100) * beta / 2

plt.figure(figsize=(15, 4))
plt.scatter(y, x, color='blue', label='feature')
plt.xlabel('y')
plt.ylabel('x')
plt.show()

data = pd.DataFrame({'target': y, 'feature': x})
```

```python
X = data['feature']
y = data['target']

model = reg.LinearRegression()
model.fit(x, y)

coefficient = model.coefficients
intercept = model.intercept

print('coefficient:', coefficient)
print('intercept:', intercept)
```
