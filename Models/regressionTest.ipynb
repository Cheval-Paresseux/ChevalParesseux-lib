```python
import LinearRegression as reg
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
beta = 2
alpha = 10

x = np.random.randn(100)
x2 = np.random.randn(100)
y = alpha + beta*x + np.random.randn(100) 

plt.figure(figsize=(15, 4))
plt.scatter(y, x, color='blue', label='feature')
plt.scatter(y, x2, color='red', label='feature2')
plt.xlabel('y')
plt.ylabel('x')
plt.show()

data = pd.DataFrame({'target': y, 'feature1': x, 'feature2': x2})
```

```python
X = data[['feature1', 'feature2']]
y = data['target']

model = reg.MSERegression()
model.fit(X, y)

coefficient = model.coefficients
intercept = model.intercept

print('coefficient:', coefficient)
print('intercept:', intercept)

model.plot_loss_history()
model.plot_fitted_line(feature_index=0)
```

```python
statistics = model.get_statistics()
print(statistics)
```

```python
residuals = model.plot_residuals()
```
